```python
import pandas as pd
```

```python
excel_file_path = "tripinfo.xlsx"
df = pd.read_excel(excel_file_path)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
def get_direction(vehicle_id):
    parts = vehicle_id.split("_")
    direction = parts[1]
    return direction
```

```python
df["Ingress Point"] = df["Vehicle_ID"].apply(get_direction)
df.head()
```

```python
directions = ["north", "south", "east", "west"]
results = {}

for direction in directions:
        # Filter the DataFrame for the current direction
        df_direction = df[df['Ingress Point'] == direction]

        # Calculate the number of vehicles
        num_vehicles = len(df_direction)

        # Calculate the metrics only if there are vehicles for the direction
        if num_vehicles > 0:
            # Calculate average travel time
            average_travel_time = df_direction['Duration'].mean()

            # Calculate average waiting time
            average_waiting_time = df_direction['Waiting Time'].mean()

            # Calculate average departure delay
            average_departure_delay = df_direction['Departure Delay'].mean()
        else:
            average_travel_time = 0
            average_waiting_time = 0
            average_departure_delay = 0

        # Store the results in the dictionary
        results[direction] = {
            'number_of_vehicles': num_vehicles,
            'average_travel_time': average_travel_time,
            'average_waiting_time': average_waiting_time,
            'average_departure_delay': average_departure_delay
        }
```

```python
df_results = pd.DataFrame(results)
df_results["Overall Average"] = df_results.iloc[:, 1:].mean(axis=1)
df_results
```
